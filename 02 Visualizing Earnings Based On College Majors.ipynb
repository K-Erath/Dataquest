Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

Rank - Rank by median earnings (the dataset is ordered by this column).  
Major_code - Major code.  
Major - Major description.  
Major_category - Category of major.  
Total - Total number of people with major.  
Sample_size - Sample size (unweighted) of full-time.  
Men - Male graduates.  
Women - Female graduates.  
ShareWomen - Women as share of total.  
Employed - Number employed.  
Median - Median salary of full-time, year-round workers.  
Low_wage_jobs - Number in low-wage service jobs.  
Full_time - Number employed 35 hours or more.   
Part_time - Number employed less than 35 hours.

# Pandas Scatter Plots
We use the kind parameter to specify the type of plot we want. We use x and y to specify the data we want on each axis.

Use scatter plots to explore the following relations:
- Sample_size and Median
- Sample_size and Unemployment_rate
- Full_time and Median
- ShareWomen and Unemployment_rate
- Men and Median
- Women and Median

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median Salary vs. Sample Size')
```

```python
recent_grads.plot(x='Sample_size',y='Unemployment_rate',kind='scatter',title='Unemployment Rate vs. Sample Size')
```

```python
recent_grads.plot(x='Full_time',y='Median',kind='scatter',title='Median Salary vs. Full Time Workers')
```

```python
recent_grads.plot(x='ShareWomen',y='Unemployment_rate',kind='scatter',title='Unemployment Rate vs. Women as Share fo Total')
```

```python
recent_grads.plot(x='Men',y='Median',kind='scatter',title='Median Salary vs. Number of Male Graduates')
```

```python
recent_grads.plot(x='Women',y='Median',kind='scatter',title='Median Salary vs. Number of Female Graduates')
```

Do students in more popular majors make more money?

```python
recent_grads.plot(x='Total',y='Median',kind='scatter',title='Median vs. Total')
```

No.

Do students that majored in subjects that were majority female make more money?

```python
recent_grads.plot(x='ShareWomen',y='Median',kind='scatter',title='Median vs. Share of Women')
```

```python
recent_grads.plot(x='Full_time',y='Median',kind='scatter',title='Median vs. Full Time Workers')
```

We can make a few observations from these scatter plots. Students in popular majors do not necessarily make more money. There is no correlation between the number of full-time employees and and median salary. There was a slight negative correlation between subjects that were majority female and median salary.

## Pandas Histograms
Generate histograms in separate jupyter notebook cells to explore the distributions of the following columns:
- Sample_size

- Full_time

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40, title=cols[r])
```
